```python
# Follow this link https://azure.microsoft.com/try/cognitive-services/?api=bing-web-search-api
# to get 7-day trial API key
API_KEY='...'
```

```python
search_url = "https://api.cognitive.microsoft.com/bing/v7.0/images/search"
headers = {"Ocp-Apim-Subscription-Key" : API_KEY}
```

```python
search_term = 'red tassel-flower'
```

```python
search_params = {"q": search_term, "license": "public", "imageType": "photo"}
```

```python
import requests
```

```python
response = requests.get(url = search_url, params = search_params, headers = headers)
```

```python
response.raise_for_status()
```

```python
results = response.json()
```

```python
# number of images found
print(len(results['value']))
```

```python
thumbnail_urls = [(image['name'], image['thumbnailUrl']) for image in results['value']]
thumbnail_urls = thumbnail_urls[:8]
```

```python
thumbnail_urls
```

```python
# display the thumbnails
import matplotlib.pyplot as plt
from matplotlib.image import imread
import urllib
```

```python
plt.figure(figsize=(20,10))
columns = 5

for i, (name,url) in enumerate(thumbnail_urls):
    f = urllib.request.urlopen(url)
    img = imread(f, format='jpg')
    plt.subplot(len(thumbnail_urls) / columns + 1, columns, i + 1)
    plt.imshow(img)
```
